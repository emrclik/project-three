## Divvy Bike Share May 2017 - Subscriber Trip Data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from datetime import *
import time
```

```python
data= pd.read_csv('chicago-divvy-bicycle-sharing-data\divvy_subscribers_cleaned.csv')
list(data.columns.values)
```

```python
#Format trip dataframe 

data = data.drop('Unnamed: 0', axis =1)
data = data.drop('TRIP ID', axis=1)
data = data.drop('START TIME', axis=1)
data = data.drop('STOP TIME', axis=1)
data = data.drop('FROM STATION NAME', axis=1)
data = data.drop('TO STATION NAME', axis=1)
data = data.drop('GENDER', axis=1)
data = data.drop('Start Date', axis=1)
data = data.drop('Start Time', axis=1)
data = data.drop('End Date', axis=1)
data = data.drop('End Time', axis=1)
data = data.drop('StartTime_Hour', axis=1)
data = data.drop('StopTime_Hour', axis=1)
data = data.drop('StartTime_Unix', axis=1)
#data = data.drop('Start_Coords', axis=1)
#data = data.drop('End_Coords', axis=1)
data = data.drop('Age', axis=1)
data = data.drop('Size', axis=1)
```

```python
data.head()
```

```python
Size = data.groupby(["Trip"]).size().sort_values(ascending=False).reset_index(name = 'Size')
Size
```

```python
data.drop_duplicates(subset="Trip", inplace = True)
data
```

```python
#Merge formatted data
trips_summary = pd.merge(data, Size, on= "Trip")
trips_summary.reset_index(drop=False)
trips_summary
```

```python
#Filter Data to Match CitiBike 
#Total of 263,666 trips in May - Subscriber Data
#Total of 43,214 trips grouped by location
#500ish bike stations
#>100 trips per location only results in 92 trips
#>50 trips to get a greater dataset (443 trips)


trips_summary = trips_summary.loc[trips_summary['Size'] > 50,:]
trips_summary_data= trips_summary.sort_values('Size', ascending=False)
trips_summary_data
```

```python
trips_summary_data["Public_Duration"] = ''
trips_summary_data.head()
```

Save trips dataframe as csv

```python
trips_summary_data.to_csv("chicago-divvy-bicycle-sharing-data\divvy_subscribers_trips.csv")
```
