## Divvy Bike Share May 2017 - Subscriber

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from datetime import *
import time
```

## Clean Data

Raw Data

```python
df3_cl = pd.read_csv("chicago-divvy-bicycle-sharing-data\Divvy_Trips_-_Subscriber-Only.csv")
```

 Beginning Cleaning Data proccess

```python
#TIMESTAMP
#Separate date and time 
df3_cl['Start Date'] = pd.to_datetime(df3_cl['START TIME']).dt.date
df3_cl['Start Time'] = pd.to_datetime(df3_cl['START TIME']).dt.time

df3_cl['End Date'] = pd.to_datetime(df3_cl['STOP TIME']).dt.date
df3_cl['End Time'] = pd.to_datetime(df3_cl['STOP TIME']).dt.time
df3_cl.dropna()


#Convert START/STOP TIME from pandas object to datetime object --> convert to unix time 
df3_cl["START TIME"] = pd.to_datetime(df3_cl["START TIME"])
df3_cl["STOP TIME"] = pd.to_datetime(df3_cl["STOP TIME"])

#Convert START/STOP TIME from pandas object to datetime object --> convert to unix time 
df3_cl["StartTime_Hour"]= df3_cl["START TIME"].dt.hour
df3_cl["StopTime_Hour"]= df3_cl["STOP TIME"].dt.hour

#Convert START time to Unix Time 
df3_cl["StartTime_Unix"] = df3_cl['START TIME'].astype(np.int64)//10**9

 


#TRIP DURATION
#Convert trip duration to int
df3_cl['TRIP DURATION (Secs)'] = df3_cl['TRIP DURATION (Secs)'].str.replace(',', '')

df3_cl["Trip_Duration"] = pd.to_numeric(df3_cl['TRIP DURATION (Secs)'], errors = 'coerce')
df3_cle = df3_cl.drop('TRIP DURATION (Secs)', axis =1)


#Convert trip duration to Minutes 
df3_cle['Trip_Duration'] = df3_cle['Trip_Duration'] / 60



#AGE
def calculate_age(row):
    return 2018 - row["BIRTH YEAR"] 
df3_cle ["Age"] = df3_cle.apply(calculate_age, axis =1)



#COODINATES
#COMBINE TRIP COORDINATES
df3_cle["Trip"] = df3_cle["FROM LATITUDE"].map(str) +  ',' + df3_cle['FROM LONGITUDE'].map(str) + '|' + df3_cle['TO LATITUDE'].map(str) + ',' + df3_cle['TO LONGITUDE'].map(str)
df3_cle["Trip"]

#COMBINE START COODINATES
df3_cle["Start_Coords"] = df3_cle["FROM LATITUDE"].map(str) +  ',' + df3_cle['FROM LONGITUDE'].map(str)
#COMBINE END COORDINATES
df3_cle["End_Coords"] = df3_cle['TO LATITUDE'].map(str) + ',' + df3_cle['TO LONGITUDE'].map(str)

df3_cle




```

```python
#MEDIAN TRIP DURATION
df3_cle['Median_Duration_Mins'] = df3_cle['Trip'].map(df3_cle.groupby('Trip')['Trip_Duration'].median())
df3_cle.head()
```

```python
#MEDIAN UNIX TIME 
df3_cle['Median_Unix_StartTime'] = df3_cle['Trip'].map(df3_cle.groupby('Trip')['StartTime_Unix'].median())
df3_cle['Median_Unix_StartTime'].head()
```

```python
#Create Size variable 
Size = pd.DataFrame(df3_cle["Trip"].value_counts())
Size.reset_index(inplace = True)
Size.columns = ["Trip", "Size"]
Size.sort_values(by = "Size", ascending= False)
Size.head()
```

```python
#Merge formatted data
df3_clea = pd.merge(df3_cle, Size, on= "Trip")
df3_clea.head()
```

```python
#MEDIAN AGE
df3_clea['Median_Age'] = df3_cle['Trip'].map(df3_clea.groupby('Trip')['Age'].median())
df3_clea
```

Clean Dataframe

```python
list(df3_clea.columns.values)
```

```python
#Clean DataFrame:
df3_clea = df3_clea.drop('BIKE ID', axis=1)
df3_clea = df3_clea.drop('USER TYPE', axis=1)
df3_clea = df3_clea.drop('BIRTH YEAR', axis=1)
df3_clea = df3_clea.drop('FROM LATITUDE', axis=1)
df3_clea = df3_clea.drop('FROM LONGITUDE', axis=1)
df3_clea = df3_clea.drop('TO LATITUDE', axis=1)
df3_clea = df3_clea.drop('TO LONGITUDE', axis=1)
df3_clea = df3_clea.drop('FROM LOCATION', axis=1)
df3_clea = df3_clea.drop('TO LOCATION', axis=1)
df3_clea = df3_clea.drop('FROM STATION ID', axis=1)
df3_clea = df3_clea.drop('TO STATION ID', axis=1)
```
