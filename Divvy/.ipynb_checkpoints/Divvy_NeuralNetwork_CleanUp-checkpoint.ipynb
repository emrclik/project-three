```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

df=  pd.read_csv('chicago-divvy-bicycle-sharing-data\divvy_subscribers_cleaned.csv')
df.head()
```

```python
#Code Gender ( 1=Male, 2= Female)
df['gender'] = df['GENDER'].apply({'Male':1, 'Female': 2}.get)
df.head()
```

```python
#list(df.columns.values)
```

```python
#Clean DataFrame:
df = df.drop('TRIP ID', axis=1)
df = df.drop('START TIME', axis=1)
df = df.drop('STOP TIME', axis=1)
df = df.drop('FROM STATION NAME', axis=1)
df = df.drop('TO STATION NAME', axis=1)
df = df.drop('GENDER', axis=1)
df = df.drop('Start Date', axis=1)
df = df.drop('Start Time', axis=1)
df = df.drop('End Date', axis=1)
df = df.drop('End Time', axis=1)
df = df.drop('StopTime_Hour', axis=1)
df = df.drop('StartTime_Unix', axis=1)
df = df.drop('Trip', axis=1)
df = df.drop('Median_Duration_Mins', axis=1)
df = df.drop('Median_Unix_StartTime', axis=1)
df = df.drop('Size', axis=1)
df = df.drop('Median_Age', axis=1)
```

```python
twenties = []
for val in df['Age']:
    if val < 30 and val > 19:
        twenties.append(1)
    else:
        twenties.append(0)
        
df['twenties'] = twenties
df.head()
```

```python
df.to_csv('chicago-divvy-bicycle-sharing-data\divvy_cleaned_neuralnetwork.csv')
```
